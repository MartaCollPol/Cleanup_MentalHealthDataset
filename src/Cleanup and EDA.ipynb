# Mental Health at Workplace:

El objetivo de este notebook es limpiar y realizar un analísis exploratorio del siguiente conjunto de datos, facilitando la extracción de conclusiones del estudio: https://public.tableau.com/app/profile/marta.coll.pol/viz/MentalHealth_16877009658200/Historia1

#### Se importan los datos:

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("..\data\survey.csv")
```

```python
df.head(10)
```

## Análisis de las columnas conflictivas:

#### Country:

```python
print(df['Country'].value_counts())
print("\n \n")
print(df['state'].unique())
```

Sería muy engañoso concluir que un determinado país enfrenta más problemas de salud mental en los empleados debido a que alrededor del 60% de las personas pertenecen a Estados Unidos. Además, hay muchos países que tienen solo un encuestado. Por lo tanto, la columna del país se vuelve insignificante. Vamos a eliminarla. Un vistazo rápido a los estados sugiere que esto solo es aplicable a los que están en Estados Unidos, por lo que también los eliminaremos.

```python
df.drop(columns=['Timestamp', 'Country', 'state', 'comments'], inplace = True)
```

#### Age:

```python
print(df['Age'].unique())
```

Se eliminan los valores imposibles:

```python
df.drop(df[df['Age'] < 0].index, inplace = True) 
df.drop(df[df['Age'] > 100].index, inplace = True)
df['Age'].unique()
```

#### Gender:

```python
print(df['Gender'].unique())
```

Se simplifican los generos en Male y Female:

```python
df['Gender'].replace(['Male ', 'male', 'M', 'm', 'Male', 'Cis Male',
                     'Man', 'cis male', 'Mail', 'Male-ish', 'Male (CIS)',
                      'Cis Man', 'msle', 'Malr', 'Mal', 'maile', 'Make',], 'Male', inplace = True)

df['Gender'].replace(['Female ', 'female', 'F', 'f', 'Woman', 'Female',
                     'femail', 'Cis Female', 'cis-female/femme', 'Femake', 'Female (cis)',
                     'woman',], 'Female', inplace = True)

df["Gender"].replace(['Female (trans)', 'queer/she/they', 'non-binary',
                     'fluid', 'queer', 'Androgyne', 'Trans-female', 'male leaning androgynous',
                      'Agender', 'A little about you', 'Nah', 'All',
                      'ostensibly male, unsure what that really means',
                      'Genderqueer', 'Enby', 'p', 'Neuter', 'something kinda male?',
                      'Guy (-ish) ^_^', 'Trans woman',], 'Other', inplace = True)

df['Gender'].value_counts()
```

Sin ser conscientes de ello, hemos descubierto que la cantidad de hombres en el conjunto de datos es cuatro veces mayor que la cantidad de mujeres. Por lo tanto, es importante tener en cuenta esta información y evitar hacer suposiciones incorrectas, como que los hombres son más propensos a tener problemas de salud mental, entre otros.

Como alternativa, podemos concluir que en la industria tecnológica hay significativamente más hombres que mujeres (este estudio se centró específicamente en la industria tecnológica).

#### Encoding de las variablas categoricas:

```python
from sklearn.preprocessing import LabelEncoder

object_cols = ['Gender', 'self_employed', 'family_history', 'treatment',
       'work_interfere', 'no_employees', 'remote_work', 'tech_company',
       'benefits', 'care_options', 'wellness_program', 'seek_help',
       'anonymity', 'leave', 'mental_health_consequence',
       'phys_health_consequence', 'coworkers', 'supervisor',
       'mental_health_interview', 'phys_health_interview',
       'mental_vs_physical', 'obs_consequence']
label_encoder = LabelEncoder()
for col in object_cols:
    label_encoder.fit(df[col])
    df[col] = label_encoder.transform(df[col])
```

#### Matriz de correlación:

El objetivo es ver que variables estan más correlacionadas para inspirar preguntas interesantes a responder con el conjunto:

```python
corr = df.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(15, 15))

sns.heatmap(corr, mask=mask, cmap='Purples', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot = True)
```

Se observa una fuerte relación entre las variablas que implican que la empresa cuenta con programas de ayuda al bienestar del empleado con la predisposición a buscar ayuda.
